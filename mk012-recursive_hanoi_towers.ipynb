## Problem: Hanoi Towers 
A number of disks is stacked in decreasing order from the bottom to the top of one rod, i.e. the largest disk at the bottom and the smallest one on top. The disks build a conical tower.    

<img align="center" src="mk012-recursive_hanoi_towers.gif" alt="Hanoi tower of 4 disks">

The aim of the game is to move the tower of disks from one rod to another rod. The following rules have to be obeyed:

1. Only one disk may be moved at a time.
2. Only the most upper disk from one of the rods can be moved in a move.
3. It can be put on another rod, if this rod is empty or if the most upper disk of this rod is larger than the one which is moved.


```python
def print_move(start, target):
    """
    This function prints the details of a disk movement whenever there is one.
    """    
    print('move ' + str(start) + ' -> ' + str(target))
    

def towers(disk_count, start, target, spare):
    """
    This function moves all but the last(the one at the bottom) disks to the spare peg, 
    moves the remaining one to the target peg and then moves again the ones at the spare peg 
    to the target peg, assuming that it is known to move an hanoi tower of one less disk. 
    Even if, it is not known the function can go to the simplest hanoi tower which is 1 
    and goes up from there, recursively.
    """    
    if disk_count == 1:
        print_move(start, target)        
    else:        
        towers(disk_count-1, start, spare, target) # moves all but the biggest to the spare
        towers(1, start, target, spare)            # moves biggest to the spare
        towers(disk_count-1, spare, target, start) # moves the ones at the spare to the target
```

```python
towers(7, "start", "target", "spare")
```
